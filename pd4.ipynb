```python
num = int(input("Enter a number: "))
is_prime = True

if num < 2:
    is_prime = False
else:
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prime = False
            break

if is_prime:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```

```python
# Function to calculate factorial using a for loop
def factorial(n):
    if n < 0:
        return "Factorial not defined for negative numbers"
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

# Function to generate Fibonacci series using a while loop
def fibonacci(n):
    fib_series = []
    a, b = 0, 1
    count = 0
    while count < n:
        fib_series.append(a)
        a, b = b, a + b
        count += 1
    return fib_series

# Function to check if a number is prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

# Main program
while True:
    print("\nChoose an operation:")
    print("1. Factorial")
    print("2. Fibonacci Series")
    print("3. Prime Number Check")
    print("4. Exit")

    choice = int(input("Enter your choice: "))

    if choice == 1:
        num = int(input("Enter a number: "))
        print(f"Factorial of {num} is {factorial(num)}")
    elif choice == 2:
        num = int(input("Enter the number of terms: "))
        print(f"Fibonacci Series: {fibonacci(num)}")
    elif choice == 3:
        num = int(input("Enter a number: "))
        if is_prime(num):
            print(f"{num} is a prime number")
        else:
            print(f"{num} is not a prime number")
    elif choice == 4:
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Please try again.")
```

```python
import getpass
users=[]
def sign_up():
    name= input("Enter your name: ")
    email=input("Enter your email: ")
    password=getpass.getpass("Enter your password: ")
    for user in users:
        if user['Email']==email:
            print("Email already registered.  Try logging in.")
            return
    
    users.append({"Name": name, "Email": email, "Password": password})
    print("Sign-up successful!")

def login():
    email = input("Enter your email: ")
    password = getpass.getpass("Enter your password: ")
    
    for user in users:
        if user['Email'] == email and user['Password'] == password:
            print("Login successful!")
            return
    print("Invalid credentials. Try again.")

def display_users():
    if not users:
        print("No users registered yet.")
        return
    
    print("\nRegistered Users:")
    print("--------------------------------")
    print("{:<20} {:<30}".format("Name", "Email"))
    print("--------------------------------")
    for user in users:
        print("{:<20} {:<30}".format(user['Name'], user['Email']))
    print("--------------------------------")

def delete_user():
    email = input("Enter the email of the user to delete: ")
    global users
    users = [user for user in users if user['Email'] != email]
    print("User deleted successfully (if existed).")

def main():
    while True:
        print("\n1. Sign Up\n2. Login\n3. Display Users\n4. Delete User\n5. Exit")
        choice = input("Choose an option: ")
        
        if choice == "1":
            sign_up()
        elif choice == "2":
            login()
        elif choice == "3":
            display_users()
        elif choice == "4":
            delete_user()
        elif choice == "5":
            print("Exiting...")
            break
        else:
            print("Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```

```python
import getpass

# User data storage
users = []

def sign_up():
    name = input("Enter your name: ")
    email = input("Enter your email: ")
    
    # Check if email is already registered
    if any(user['Email'] == email for user in users):
        print("Email already registered. Try logging in.")
        return
    
    password = getpass.getpass("Enter your password: ")
    confirm_password = getpass.getpass("Confirm your password: ")
    
    if password != confirm_password:
        print("Passwords do not match. Please try again.")
        return
    
    users.append({"Name": name, "Email": email, "Password": password})
    print("Sign-up successful!")

def login():
    email = input("Enter your email: ")
    password = getpass.getpass("Enter your password: ")
    
    for user in users:
        if user['Email'] == email and user['Password'] == password:
            print("Login successful!")
            return
    print("Invalid credentials. Try again.")

def display_users():
    if not users:
        print("No users registered yet.")
        return
    
    print("\nRegistered Users:")
    print("--------------------------------")
    print("{:<20} {:<30}".format("Name", "Email"))
    print("--------------------------------")
    for user in users:
        print("{:<20} {:<30}".format(user['Name'], user['Email']))
    print("--------------------------------")

def delete_user():
    email = input("Enter the email of the user to delete: ")
    global users
    
    if any(user['Email'] == email for user in users):
        users = [user for user in users if user['Email'] != email]
        print(f"User with email {email} has been deleted.")
    else:
        print("No user found with this email.")

def main():
    while True:
        print("\n1. Sign Up\n2. Login\n3. Display Users\n4. Delete User\n5. Exit")
        choice = input("Choose an option: ")
        
        if choice == "1":
            sign_up()
        elif choice == "2":
            login()
        elif choice == "3":
            display_users()
        elif choice == "4":
            delete_user()
        elif choice == "5":
            print("Exiting...")
            break
        else:
            print("Invalid choice. Try again.")

if __name__ == "__main__":
    main()
```
